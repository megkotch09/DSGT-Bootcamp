# Introductory Data Preprocessing
Data Preprocessing is a vital component of data science. This exercise will allow you to get exposure to preprocessing a dataset and what things to consider when analyzing your datasets. 

In `exercise.py`, you will find a set of tasks to complete in order to ramp up your foundations. There is also a file called `tester.py` that you can run in order to see if your implementation works (through unit tests). Couple pieces of advice:

1. Take your time with this exercise. You want to make sure that you clearly understand the problem that is presented

2. Don't just copy solutions off of the Internet blindly. You can use resources such as Stack Overflow as helpers, but don't substitute the actual learning just for "completing the exercise". Your goal is to sharpen your problem solving

3. Collaborate at the "whiteboard-level" (like discuss approaches at high level/bounce ideas, but not exactly specific implementation details)

4. Work with other students in the Udemy Course (we have a Slack channel after all that you joined). Please refrain from sharing your solutions to the group. We would hate if the meat and potatoes of the problems got "spoiled". 

Good luck in this exercise!
